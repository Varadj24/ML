```python
#Q1.Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
```

```python
'''
1).Histogram:
Use: Histograms are used to visualize the distribution of a single numeric variable. They are great for understanding the underlying frequency distribution of the data, including information about the central tendency and spread of the variable.
Seaborn Function: sns.histplot(data, bins=10)

2).Scatter Plot:
Use: Scatter plots are used to display the relationship between two numeric variables. They are useful for identifying patterns, correlations, clusters, or outliers in the data.
Seaborn Function: sns.scatterplot(x="x_variable", y="y_variable", data=data)

3).Bar Plot:
Use: Bar plots are used to represent categorical data with rectangular bars. They are useful for comparing the quantities of different categories. Bar plots can be customized to display counts, percentages, or other aggregated values.
Seaborn Function: sns.barplot(x="category", y="value", data=data)

4).Box Plot:
Use: Box plots (or box-and-whisker plots) are used to visualize the distribution of a numeric variable across different categories. They provide insights into the data's central tendency, spread, and presence of outliers.
Seaborn Function: sns.boxplot(x="category", y="numeric_variable", data=data)

5).Heatmap:
Use: Heatmaps are used to display a matrix of data in a grid, where colors represent the values in the matrix. They are particularly useful for displaying correlations in a dataset, highlighting patterns, or clustering of variables.
Seaborn Function: sns.heatmap(data, annot=True, cmap="coolwarm")'''
```

```python
#Q2.Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions. 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fmri_data = sns.load_dataset("fmri")

plt.figure(figsize=(10, 6))
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)

plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("FMRI Signal over Time for Different Events and Regions")

plt.legend(title="Event")
plt.show()
```

```python
#Q3. Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
```

```python
titanic_data = sns.load_dataset("titanic")

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.xlabel('Pclass')
plt.ylabel('Age')
plt.title('Pclass vs. Age')

plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.title('Pclass vs. Fare')

plt.tight_layout()
plt.show()
```

```python
#Q4.Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.
```

```python
diamonds_data = sns.load_dataset("diamonds")

plt.figure(figsize=(8, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', multiple="stack", palette="Set1", kde=True)
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Histogram of Prices by Cut')
plt.show()
```

```python
#Q5. Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.
```

```python
iris_data = sns.load_dataset("iris")

sns.pairplot(data=iris_data, hue='species', palette='Set1')

plt.show()
```

```python
#Q6.Use the "flights" dataset from seaborn to plot a heatmap.
```

```python
flights_data = sns.load_dataset("flights")

flights_pivot = flights_data.pivot("month", "year", "passengers")

plt.figure(figsize=(10, 8))
sns.heatmap(data=flights_pivot, annot=True, fmt="d", cmap="YlGnBu", linewidths=1, cbar_kws={'label': 'Passengers'})
plt.title("Passenger Counts (1000s) by Year and Month")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
